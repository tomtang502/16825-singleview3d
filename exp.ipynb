```python
import torch, torchvision
import os
print(os.cpu_count())

import dataset_location
from r2n2_custom import R2N2
from  pytorch3d.datasets.r2n2.utils import collate_batched_R2N2
import matplotlib.pyplot as plt
import numpy as np
```

```python
sample_latent = torch.load("writeup_assets/2/2.6/350_point_dig.pt")
```

```python
sample_latent.keys()
```

```python
sample_latent['dconv0'][0].shape
```

```python
# Create a figure and 16x16 subplots
fig, axes = plt.subplots(nrows=16, ncols=16, figsize=(10, 10))

for i in range(256):
    row = i // 16
    col = i % 16
    
    # Plot the 5x5 channel as a grayscale image
    axes[row, col].imshow(sample_latent['dconv0'][0][i].detach().cpu().numpy(), cmap='viridis')
    axes[row, col].axis('off')

plt.tight_layout
plt.savefig("writeup_assets/2/2.6/firstdconv.png", dpi=300)
```

```python

nrows, ncols = 16, 8  # 16 * 8 = 128
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(10, 20))

for i in range(128):
    row = i // ncols
    col = i % ncols
    axes[row, col].imshow(sample_latent['decoder_conv2d_relu_1'][0][i].detach().cpu().numpy(), cmap='viridis')
    axes[row, col].axis('off')

plt.tight_layout()
plt.savefig("writeup_assets/2/2.6/layer0.png", dpi=300)
plt.show()
```

```python

nrows, ncols = 8, 8  # 8*8=64
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(10, 10))

for i in range(64):
    row = i // ncols
    col = i % ncols
    axes[row, col].imshow(sample_latent['decoder_conv2d_relu_2'][0][i].detach().cpu().numpy(), cmap='viridis')
    axes[row, col].axis('off')

plt.tight_layout()
plt.savefig("writeup_assets/2/2.6/layer1.png", dpi=300)
plt.show()
```

```python
nrows, ncols = 4, 8  # 4 * 8 = 32
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, 6))

for i in range(32):
    row = i // ncols
    col = i % ncols
    axes[row, col].imshow(sample_latent['dconv_relu_1'][0][i].detach().cpu().numpy(), cmap='viridis')
    axes[row, col].axis('off')

plt.tight_layout()
plt.savefig("writeup_assets/2/2.6/layer2.png", dpi=300)
plt.show()
```

```python
nrows, ncols = 4, 4  # 4 * 4 = 16
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(8, 8))

for i in range(16):
    row = i // ncols
    col = i % ncols
    axes[row, col].imshow(sample_latent['dconv_relu_3'][0][i].detach().cpu().numpy(), cmap='viridis')
    axes[row, col].axis('off')

plt.tight_layout()
plt.savefig("writeup_assets/2/2.6/layer3.png", dpi=300)
plt.show()
```

```python

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
channel_labels = ['x', 'y', 'z']

for i in range(3):
    axes[i].imshow(sample_latent['xyz_res'][0][i].detach().cpu().numpy(), cmap='viridis')  
    axes[i].set_title(channel_labels[i])
    axes[i].axis('off')


plt.tight_layout()

plt.savefig("writeup_assets/2/2.6/three_channels.png", dpi=300)
plt.show()
```
